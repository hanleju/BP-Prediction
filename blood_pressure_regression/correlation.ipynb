```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = './otherdata/our_bmi.csv'

df = pd.read_csv(file_path)

correlation_matrix = df.corr()

selected_columns = ['SBP','DBP']
selected_correlation_matrix = correlation_matrix[selected_columns]

print(selected_correlation_matrix)

plt.figure(figsize=(10, 20))
plt.rc('font',size=20)
sns.heatmap(selected_correlation_matrix, cmap='coolwarm', annot=True, fmt=".4f")
plt.title('feature correlation matrix')
plt.show()
```
