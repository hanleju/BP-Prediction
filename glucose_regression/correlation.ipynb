```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = './glucose_data/train_vital_data.csv'
df = pd.read_csv(file_path, encoding='utf-8')
df = df.drop('time', axis=1)
df['bmi'] = df['weight']/((df['height']/100)**2)
df = df.dropna()

df.head()
```

```python
correlation_matrix = df.corr()
selected_colums = ["glucose"]
selected_correlation_matrix = correlation_matrix[selected_colums]

print(selected_correlation_matrix)
plt.figure(figsize=(8,12))
plt.rc('font',size=15)
sns.heatmap(selected_correlation_matrix, cmap='coolwarm', annot=True, fmt='.3f')
plt.title('glucose correlation')
plt.show()
```
